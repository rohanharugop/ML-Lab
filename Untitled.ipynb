```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Datasets/PlayTennis.csv")
df.head()
```

```python
X = df.drop( "play" , axis=1)
y = df["play"]
```

```python
label_encoder = {}
for cols in X.columns:
    le = LabelEncoder()
    X[cols] = le.fit_transform(X[cols])
    label_encoder[cols] = le
```

```python
tle = LabelEncoder()
y = tle.fit_transform(y)
```

```python
clf = DecisionTreeClassifier(criterion="entropy" , random_state=42)
clf.fit(X,y)
plt.figure(figsize=(12,8))
plot_tree(clf,feature_names=X.columns, class_names=tle.classes_, filled=True, fontsize=11)
```

```python
sample_datapoint = {
"outlook": "sunny",
"temp": "cool",
"humidity": "high",
"windy": "strong"
}
```

```python
sample_df = pd.DataFrame([sample_datapoint])
```

```python
for col, le in label_encoder.items():
    sample_df[col] = le.fit_transform(sample_df[col])
    
```

```python
pec = clf.predict(sample_df)
```

```python
pred_class = tle.inverse_transform(pec)
```

```python
print(f"Predicted Class: {pred_class}")
```
